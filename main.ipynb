# EDA

```python
weather_condition = df['weather_condition'].value_counts().sort_index()

weather_condition.plot(kind='bar', color='red', edgecolor='black')
plt.title('Weather Condition')
plt.xlabel('Keadaan Cuaca')
plt.ylabel('Jumlah Kejadian')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

clear = df[df['weather_condition'] == 'CLEAR'].shape[0]
rain = df[df['weather_condition'] == 'RAIN'].shape[0]
snow = df[df['weather_condition'] == 'SNOW'].shape[0]
cloudy_or_overcast = df[df['weather_condition'] == 'CLOUDY/OVERCAST'].shape[0]
unknown = df[df['weather_condition'] == 'UNKNOWN'].shape[0]
fog_or_smoke_or_haze = df[df['weather_condition'] == 'FOG/SMOKE/HAZE'].shape[0]
blowing_snow = df[df['weather_condition'] == 'BLOWING SNOW'].shape[0]
freezing_rain_or_drizzle = df[df['weather_condition'] == 'FREEZING RAIN/DRIZZLE'].shape[0]
other = df[df['weather_condition'] == 'OTHER'].shape[0]
sleet_or_hail = df[df['weather_condition'] == 'SLEET/HAI'].shape[0]
severe_cross_wind_gate = df[df['weather_condition'] == 'SEVERE CROSS WIND GATE'].shape[0]
blowing_sand_soil_dirt = df[df['weather_condition'] == 'BLOWING SAND, SOIL, DIRT'].shape[0]

data = {
    'Weather_Condition' : ['CLEAR', 'RAIN', 'SNOW', 'CLOUDY/OVERCAST', 'UNKNOWN', 'FOG/SMOKE/HAZE', 'BLOWING SNOW', 'FREEZING RAIN/DRIZZLE', 'OTHER', 'SLEET/HAIL', 'SEVERE CROSS WIND GATE', 'BLOWING SAND, SOIL, DIRT'],
    'Count' : [clear, rain, snow, cloudy_or_overcast, unknown, fog_or_smoke_or_haze, blowing_snow, freezing_rain_or_drizzle, other, sleet_or_hail, severe_cross_wind_gate, blowing_sand_soil_dirt]
}

df_weather_condition = pd.DataFrame(data)
plt.figure(figsize=(8, 8))
plt.pie(df_weather_condition['Count'], labels=df_weather_condition['Weather_Condition'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)

plt.title('Persentase Kecelakaan Berdasarkan Kondisi Cuaca')

plt.show()
```

Ini bisa terjadi mungkin dikarenakan jumlah pengendara pada saat cuaca CLEAR berpuluh kali lipat dibandingkan jumlah pengendara saat kondisi cuaca selain CLEAR

Saya hanya akan membedah kejadian pada saat cuaca 'CLEAR', 'RAIN', 'SNOW', 'CLOUDY/OVERCAST' saja karena hanya itu yang memiliki angka kejadian yang cukup banyak.

```python
# Menghitung total korban tiap cuaca

injuries_clear = df[(df['weather_condition'] == 'CLEAR') & (df['injuries_total'] > 0)]['injuries_total'].sum()
injuries_rain = df[(df['weather_condition'] == 'RAIN') & (df['injuries_total'] > 0)]['injuries_total'].sum()
injuries_snow = df[(df['weather_condition'] == 'SNOW') & (df['injuries_total'] > 0)]['injuries_total'].sum()
injuries_cloudy = df[(df['weather_condition'] == 'CLOUDY/OVERCAST') & (df['injuries_total'] > 0)]['injuries_total'].sum()

data_korban_full = {
    'cuaca' : ['clear', 'rain', 'snow', 'cloudy'],
    'total' : [injuries_clear, injuries_rain, injuries_snow, injuries_cloudy]
}

df_total_korban = pd.DataFrame(data_korban_full)

# Mencetak barplot
plt.figure(figsize=(8, 6))
plt.bar(df_total_korban['cuaca'], df_total_korban['total'], color='red', edgecolor='black')
plt.title('Jumlah Korban Kecelakaan Berdasarkan Kondisi Cuaca')
plt.xlabel('Keadaan Cuaca')
plt.ylabel('Total Korban')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Mencetak piechart
plt.figure(figsize=(8, 8))
plt.pie(df_total_korban['total'], labels=df_total_korban['cuaca'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Persentase Korban Berdasarkan Kondisi Cuaca')
plt.show()
```

Insight dari data yang telah dibedah ini menunjukan bahwa cuaca yang baik belum tentu dapat meminimalisir jumlah kecelakaan yang terjadi.
